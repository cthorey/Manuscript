```python
from wordcloud import WordCloud, STOPWORDS
Root = "/Users/thorey/Documents/These/Manuscript/"
chap1 = os.path.join(Root,'Chapter','Chapter1.tex')
chap2 = os.path.join(Root,'Chapter','Chapter2.tex')
chap3 = os.path.join(Root,'Chapter','Chapter3.tex')
chap4 = os.path.join(Root,'Chapter','Chapter4.tex')
chap5 = os.path.join(Root,'Chapter','Chapter5.tex')
chap6 = os.path.join(Root,'Chapter','Chapter6.tex')
Marion = os.path.join(Root,'main_Marion.tex')
def WordCloudT(Chap,NameWC):
    word_final = []
    for chap in Chap:
        text = open(chap,'r')
        tokenizer = RegexpTokenizer(r'\w+')
        FilterWord = ['citep','citet','begin','end','frac','partial','Unit']
        FilterWord += ['h_0','Lambda','eta_h','Figure','label','ref','left','right']
        FilterWord += ['Leur','unit','t']
        FilterWord += ['page','Thesis','bibliography','TABLES','appendices','marion','guillaume','CONTENT','LIST']
        FilterWord += ['large','Roman','symbols','CHAPTER','Contents','list','CHAPTERS','thesis','Latex','Users']
        FilterWord += ['ABBREVIATIONS','LaTeX','Tables','table','FIGURES','PACKAGE','hrule','Submited','content']
        FilterWord += ['write','Set','PACKAGES','Doctoral','Avant','logo','chapter','Template','template','APPENDICES']
        FilterWord += ['vfill','chapters','reference','pages','Define','setting','delta','represent','dominate','follow']
        FilterWord += ['enhance','cool','affect','depend','finally','variable','DELTA','figure','leur','result','section']
        FilterWord += ['tables','contents','latex']
        word = [f for f in tokenizer.tokenize(text.read()) if len(f)>2]
        wordf = filter(lambda x:x.lower() not in FilterWord,word)
        wordf = filter(lambda x:x.isalpha(),wordf)
        word_final += wordf
    print word_final.count('viscosity')
    fo = open(os.path.join(Root,'word_tmp.txt'), "w")
    fo.write("\n".join(list(set(word_final))))
    fo.close()
    
    mask = imread(path.join(Root,'Figure','WordCloud',"cerveau.jpg"))
    wc = WordCloud(background_color="white", max_words=2000, mask=mask,
                   stopwords=STOPWORDS.add("chocolat"))
    # generate word cloud
    Input = open(os.path.join(Root,'word_tmp.txt')).read()
    wc.generate(Input)
    # store to file
    wc.to_file(os.path.join(Root,NameWC))
    return wc
```

```python
Im = WordCloudT([Marion],'Marion.png')
plt.figure(figsize=(50,50))
plt.imshow(Im)
plt.axis('off')
```

```python
import sys, nltk
from wordcloud import WordCloud, STOPWORDS
Root = "/Users/thorey/Documents/These/Manuscript/"

def WordCloud2(Wordfile,NameWC):
    print Wordfile
    f = open(Wordfile, 'rU')
    txt = f.read()
    f.close()

    tokens = nltk.word_tokenize(txt) # tokenize text
    clean_tokens = []

    for word in tokens:
        word = word.lower()
        if word.isalpha(): # drop all non-words
            clean_tokens.append(word)
            
    fo = open(os.path.join(Root,'word_tmp.txt'), "w")
    fo.write("\n".join(list(set(clean_tokens))))
    fo.close()
    
    mask = imread(path.join(Root,'Figure','WordCloud',"Boule.jpg"))
    wc = WordCloud(background_color="white", max_words=2000, mask=mask,
                   stopwords=STOPWORDS.add("chocolat"))
    # generate word cloud
    Input = open(os.path.join(Root,'word_tmp.txt')).read()
    wc.generate(Input)
    # store to file
    wc.to_file(os.path.join(Root,NameWC))
    return wc
    
```

```python
Im = WordCloud2(os.path.join(Root,'main.txt'),'Cooling.png')
plt.figure(figsize=(10,10))
plt.imshow(Im)
plt.axis('off')
```
